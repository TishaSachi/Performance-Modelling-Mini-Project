```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your dataset (replace with your file path)
data = pd.read_csv("dataset.csv")

# Preview the data
print(data.head())
```

```python

# ------------------------------------ 1. Average Start Delay by Network Type -----------------------------------

avg_start_delay = data.groupby('Network Type')['Start Delay (s)'].mean()
print("\nAverage Start Delay by Network Type:")
print(avg_start_delay)

# Plot it
avg_start_delay.plot(kind='bar', color='skyblue', title='Average Start Delay by Network Type')
plt.ylabel('Start Delay (s)')
plt.xlabel('Network Type')
plt.show()
```

```python

# ----------------------------- 2. Total Buffer Time by Network Type -----------------------------------

# Replace missing or NaN with 0
data['Total Buffer Time (s)'] = pd.to_numeric(data['Total Buffer Time (s)'], errors='coerce').fillna(0)

buffer_time = data.groupby('Network Type')['Total Buffer Time (s)'].sum()
print("\nTotal Buffer Time by Network Type:")
print(buffer_time)

# Plot
buffer_time.plot(kind='bar', color='salmon', title='Total Buffer Time by Network Type')
plt.ylabel('Total Buffer Time (s)')
plt.xlabel('Network Type')
plt.show()
```

```python

# ---------------------- 3. Start Delay vs Network Speed ---------------------------------

plt.scatter(data['Speed (Mbps)'], data['Start Delay (s)'], color='green')
plt.title('Start Delay vs Network Speed')
plt.xlabel('Network Speed (Mbps)')
plt.ylabel('Start Delay (s)')
plt.show()
```

```python
# ------------------------------ 4. Buffer Count vs Network Type -------------------------------

buffer_count = data.groupby('Network Type')['Buffer Count'].sum()
print("\nBuffer Count by Network Type:")
print(buffer_count)

# Plot
buffer_count.plot(kind='bar', color='purple', title='Buffer Count by Network Type')
plt.ylabel('Buffer Count')
plt.xlabel('Network Type')
plt.show()
```

```python
# --------------------------- 5. Optional: Start Delay by Time of Day -------------------------------

avg_delay_time = data.groupby('Time of Day')['Start Delay (s)'].mean()
print("\nAverage Start Delay by Time of Day:")
print(avg_delay_time)

avg_delay_time.plot(kind='bar', color='orange', title='Average Start Delay by Time of Day')
plt.ylabel('Start Delay (s)')
plt.xlabel('Time of Day')
plt.show()
```
